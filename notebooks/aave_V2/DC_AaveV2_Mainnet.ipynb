```python
import glob
import pandas as pd
import numpy as np


def read_file(fn):

    df = pd.read_csv(fn)

    df["timestamp"] = fn.split("/")[-1].split("_")[0]

    return df


def combine_like_files(fn, func):

    f_names = glob.glob(fn)

    df = pd.concat(map(func, f_names)).reset_index(drop=True)

    return df

df = combine_like_files(str(path/f"DataGathering/aaveV2/aave_mainnet/*.csv"),read_file)
```

# Cleaning

```python
df_copy = df.copy()

def eval_col(df,cols):

    for col in cols:
        df[col] = df[col].apply(lambda x: eval(x))

    return df

df_cleaned = eval_col(df_copy,["price","pool"])

p = pd.json_normalize(df_copy["price"])
pp = pd.json_normalize(df_copy["pool"]).rename(columns={"id":"pool_id"})
df_cleaned = pd.concat([df_copy,p,pp],axis=1).drop(columns=["price","pool"])
df_cleaned["timestamp"] = df_cleaned["timestamp"].astype(float).astype(int)
```

# Calculations

```python
RAY = 10 ** 27 #int with 27 digits
WAD = 10 ** 18 #int with 18 digits
SECONDS_PER_YEAR = 31536000


wad_cols = ["aEmissionPerSecond","vEmissionPerSecond","priceInEth"]

ray_cols = ["variableBorrowRate","liquidityRate"]

token_decimals = ["totalATokenSupply","totalCurrentVariableDebt","totalLiquidity","availableLiquidity"]

annualize_cols = ["aEmissionPerSecond","vEmissionPerSecond"]

cols_convert = wad_cols + ray_cols + token_decimals + annualize_cols

for col in cols_convert:

     df_cleaned[col] = df_cleaned[col].astype(float)
```

```python
df_cleaned.dtypes
```

```python

def calc_APR_APY(df,cols):

    for col in cols:

        col_name_APR = "_".join([col,"APR"])

        col_name_APY = "_".join([col,"APY"])

        df[col_name_APR] =  df[col].astype(float) / RAY

        df[col_name_APY] =  ((1 + (df[col_name_APR] / SECONDS_PER_YEAR)) ** SECONDS_PER_YEAR) - 1

    return df

def convert_WAD(df,cols):

    for col in cols:

        df[col] = df[col] / WAD

    return df

def convert_TokenDecimals(df,cols):

    for col in cols:

        df[col] = df[col] / 10 ** df["decimals"]
       
    return df

def annualize(df,cols):

    for col in cols:

        df[col] = SECONDS_PER_YEAR * df[col] / WAD

    return df
```

# Calculation Pipe

```python
select_tokens = ["TUSD","USDC","DAI","USDT","GUSD","WETH"]

df_st = (df_cleaned.query('''
                        symbol == "TUSD" | symbol == "USDC" | \
                        symbol == "DAI" | symbol == "USDT" | \
                        symbol == "GUSD" | symbol == "WETH" 
                        '''
                        ).sort_values("timestamp",ascending=True).reset_index(drop=True)

)
```

```python
df_st.dtypes
```

```python
#Carries each function out and returns a df
df_converted = (df_st
                .pipe(convert_WAD, cols=wad_cols)
                .pipe(calc_APR_APY,cols=ray_cols)
                .pipe(convert_TokenDecimals,cols=token_decimals)
                .pipe(annualize,annualize_cols)
)
```

```python
import time
import datetime
  
def _define_date_range(start,end):

    start_date = datetime.datetime.strptime(start,"%m/%d/%Y").timetuple()
    end_date = datetime.datetime.strptime(end,"%m/%d/%Y").timetuple()

    return time.mktime(start_date),time.mktime(end_date)

def make_datasets(df,start,end):

    start_timestamp, end_timestamp = _define_date_range(start,end) 

    return df_st.query(f'timestamp >= {start_timestamp} & timestamp <= {end_timestamp}')
```

```python
training_ds = make_datasets(df_converted,"01/01/2021","02/28/2022")
prod_mar_22 = make_datasets(df_converted,"03/01/2022","03/31/2022")
prod_apr_22 = make_datasets(df_converted,"04/01/2022","04/30/2022")
prod_may_22 = make_datasets(df_converted,"05/01/2022","05/31/2022")
```
